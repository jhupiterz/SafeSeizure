```python
import numpy as np
import pandas as pd
import pywt
import seaborn as sns
import scaleogram as scg 
import matplotlib.pyplot as plt
import matplotlib.gridspec as GridSpec
from scipy.fftpack import fft
```

```python
from sklearn.preprocessing import RobustScaler, MinMaxScaler
def scaler(df):
    return MinMaxScaler().fit_transform(df)
```

```python
data = pd.read_csv('Patient_1_csv/preictal_segment_1.csv')
```

```python
scg.set_default_wavelet('mexh')

signal_length = 5000
# range of scales to perform the transform
scales = scg.periods2scales( np.arange(1, signal_length+1) )
  
ax = scg.cws(data.iloc[0][20000:25000], scales=scales, figsize=(10, 4.0), coi = False, cmap='binary', 
        xlabel='', ylabel='', cbar='', title='')
ax.set_xticks([]);
ax.set_yticks([]);
ax.figure.set_size_inches(5, 5)
plt.savefig('test', format='png', bbox_inches='tight',pad_inches = 0)
```
